```python
import numpy as np
from sklearn.cluster import KMeans
from scipy import misc
import matplotlib.pyplot as plt
```

```python
# Load raccoon face image in grayscale
face = misc.face(gray=True)

# Display original image
plt.figure(figsize=(6, 6))
plt.imshow(face, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()

# Reshape the image data into a 2D array of pixels
# Each pixel has a single intensity value (grayscale)
pixels = face.reshape(-1, 1)

# Apply KMeans clustering to compress the image into 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(pixels)
compressed_pixels = kmeans.cluster_centers_[kmeans.labels_]

# Reshape the compressed data to the original image shape
compressed_image = compressed_pixels.reshape(face.shape)

# Display compressed image
plt.figure(figsize=(6, 6))
plt.imshow(compressed_image, cmap='gray')
plt.title("Compressed Image with 5 Clusters")
plt.axis('off')
plt.show()
```
